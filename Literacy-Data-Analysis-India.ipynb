```python
import pandas as pd

df=pd.read_csv("C:/Users/hp/Downloads/GOI.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns = [
    'Category',
    'State',
    'Total_2001',
    'Total_2011',
    'Rural_2001',
    'Rural_2011',
    'Urban_2001',
    'Urban_2011'
]

df.head()
```

```python
df=df[df['Category']=='State']
```

```python
df=df.drop(columns=['Category'])
```

```python
df['Improvement_Total']=df['Total_2011']-df['Total_2001']
df['Improvement_Rural']=df['Rural_2011']-df['Rural_2001']
df['Improvement_Urban']=df['Urban_2011']-df['Urban_2001']
```

```python
import matplotlib.pyplot as plt

df_sorted = df.sort_values(by='Improvement_Total', ascending=False)

plt.figure()
plt.bar(df_sorted['State'], df_sorted['Improvement_Total'])
plt.xticks(rotation=90)
plt.title("Literacy Improvement (2001–2011)")
plt.show()
```

```python
#Urban areas consistently show higher literacy rates compared to rural areas, indicating a persistent rural-urban gap in education.
```

```python
import matplotlib.pyplot as plt

top_states = df.sort_values(by='Total_2011', ascending=False).head(10)

plt.figure()
plt.bar(top_states['State'], top_states['Total_2011'])
plt.xticks(rotation=45)
plt.title("Top 10 States by Literacy Rate (2011)")
plt.show()
```

```python
df_sorted = df.sort_values(by='Improvement_Total', ascending=False)

plt.figure()
plt.bar(df_sorted['State'], df_sorted['Improvement_Total'])
plt.xticks(rotation=90)
plt.title("Improvement in Literacy Rate (2001–2011)")
plt.show()
```

```python
states = df['State']

plt.figure()
plt.plot(states, df['Total_2001'], label='2001')
plt.plot(states, df['Total_2011'], label='2011')
plt.xticks(rotation=90)
plt.legend()
plt.title("Literacy Rate Comparison")
plt.show()
```

```python
top_states = df.sort_values(by='Total_2011', ascending=False).head(10)

plt.figure()
plt.plot(top_states['State'], top_states['Total_2001'], label='2001')
plt.plot(top_states['State'], top_states['Total_2011'], label='2011')
plt.xticks(rotation=45)
plt.legend()
plt.title("Top 10 States Literacy Comparison")
plt.show()
```

```python
df['Gap_2011'] = df['Urban_2011'] - df['Rural_2011']

df_sorted = df.sort_values(by='Gap_2011', ascending=False)

plt.figure()
plt.bar(df_sorted['State'], df_sorted['Gap_2011'])
plt.xticks(rotation=90)
plt.title("Urban-Rural Literacy Gap (2011)")
plt.show()
```

```python
#A significant gap exists between urban and rural literacy rates across many states, highlighting inequality in access to education.
```
